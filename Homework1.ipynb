```python
sir1 = "Ana are mere" 
sir2 = "Hello Python" 
sir3 = "Pizza Party"
rsir1 = sir1.replace(" ", "_")
rsir2 = sir2.replace(" ", "_")
rsir3 = sir3.replace(" ", "_")
rsir1 = rsir1 + "."
rsir2 = rsir2 + "."
rsir3 = rsir3 + "."
csir1 = sir1.split()[0]
csir2 = sir2.split()[0]
csir3 = sir3.split()[0]
fsir1 = (csir1 + "_") * 3 + rsir1
fsir2 = (csir2 + "_") * 3 + rsir2
fsir3 = (csir3 + "_") * 3 + rsir3
print(f"{fsir1} \n{fsir2} \n{fsir3}")
```
