```python
list1 = ['abc', 123, '1', 1]
integers = floats = strings = 0
for i in list1:
    print(type(i))
    if type(i) == int:
        integers += 1
    elif type(i) == float:
        floats += 1
    elif type(i) == str:
        strings += 1
print(f"Number of int values: {integers}. \nNumber of float values: {floats}. \nNumber of strings: {strings}.")
```

```python
obiecte = ["Masa", 5, "Scaun", 4.5, [5, 6, 7], 8]
for i in obiecte:
    if type(i) == int:
        print(f"Tipul obiectului {i} este int.")
    elif type(i) == float:
        print(f"Tipul obiectului {i} este float.")
    elif type(i) == str:
        print(f"Tipul obiectului {i} este str.")
    elif type(i) == list:
        print(f"Tipul obiectului {i} este list.")
    elif type(i) == set:
        print(f"Tipul obiectului {i} este set.")
    elif type(i) == tuple:
        print(f"Tipul obiectului {i} este tuple.")
    elif type(i) == dict:
        print(f"Tipul obiectului {i} este dict.")
    elif type(i) == bool:
        print(f"Tipul obiectului {i} este bool.")
    else:
        print("Obiect de tip invalid.") 
```

```python
lista = input ("Introduceti lista de task-uri separate prin virgula: ")
lista_taskuri = lista.split(",")
print(lista_taskuri)
for i in range(0, len(lista_taskuri) - 1):
    j = i + 1
    while(j < len(lista_taskuri)):
        if lista_taskuri[i] == lista_taskuri[j]:
            lista_taskuri.pop(j)
        else:
            j += 1
print(lista_taskuri)
```

```python
sir_valori = input("Introduceti un sir de valori separate prin virgula, fara spatii, ghilimele sau apostroafe")
lista_numere_sir = sir_valori.split(",")
lista_numere = list()
print(lista_numere_sir)
i = 0
while (i < len(lista_numere_sir)):
    if not lista_numere_sir[i][0].isdigit():
        lista_numere_sir.pop(i)
    else:
        i += 1
print(lista_numere_sir)
for i in range(0, len(lista_numere_sir)):
    numar = int(lista_numere_sir[i])
    if numar >= 0 and numar <= 100:
        lista_numere.append(numar)
print(lista_numere)
print("1. Fiecare numar la puterea y\n2. Suma tuturor numerelor din lista \n3. Inmultirea fiecarui numar cu y\n4. Iesire")
index = int(input("Introduceti un index"))
while True:
    if index == 1:
        y = int(input("Introduceti un numar"))
        for i in range(0, len(lista_numere)):
            lista_numere[i] = lista_numere[i] ** y
        print(lista_numere)
    elif index == 2:
        suma = 0
        for i in range(0, len(lista_numere)):
            suma += lista_numere[i]
        print(suma)
    elif index == 3:
        y = int(input("Introduceti un numar"))
        for i in range(0, len(lista_numere)):
            lista_numere[i] = lista_numere[i] * y
        print(lista_numere)
    else:
        break
    index = int(input("Introduceti un numar"))
```

```python
numere_alese = list()
numere_castigatoare = [4, 12, 31, 17, 22, 25]
numar_potriviri = 0
for i in range(0, 6):
    numar_sir = input("Introduceti un numar")
    while(numar_sir[0].isalpha() or (int(numar_sir) < 1 or int(numar_sir) > 49) or int(numar_sir) in numere_alese):
        numar_sir = input("Valoare invalida. Introduceti un numar")
    numere_alese.append(int(numar_sir))
print(numere_alese)
for i in range(0, len(numere_alese)):
    for j in range(0, len(numere_castigatoare)):
        if numere_alese[i] == numere_castigatoare[j]:
            numar_potriviri += 1
if numar_potriviri == 2:
    print("Ati castigat 50 lei.")
elif numar_potriviri >= 3 and numar_potriviri <= 4:
    print("Ati castigat 500 lei.")
elif numar_potriviri == 5:
    print("Ati castigat 1000 lei.")
elif numar_potriviri == 6:
    print("Ati castigat 5000 lei.")
```

```python
import functools
import math
promovati = 0
studenti = {}
nume_student = input("Introduceti un nume sau STOP pentru a finaliza executia programului")
while(nume_student != "STOP"):
    lista_note = list()
    nota = int(input("Introduceti un numar de la 1 la 10, sau 0 pentru a termina lista notelor"))
    while(nota != 0):
        lista_note.append(nota)
        nota = int(input("Introduceti un numar de la 1 la 10, sau o pentru a termina lista de note"))
    studenti.update({nume_student: lista_note})
    nume_student = input("Introduceti un nume sau STOP pentru a finaliza executia programului")
for student in studenti:
    note = studenti[student]
    media = functools.reduce(lambda x, y: (x + y)/2, note)      
    studenti[student].append(media)
print(studenti)
student_max = max(studenti, key = lambda x: studenti[x][-1])
print(f"Cel mai bun student este {student_max}.")
for student in studenti:
    if studenti[student][len(studenti[student])-1] >= 6:
        promovati += 1
print(f"Au promovat {promovati} studenti.")
```
