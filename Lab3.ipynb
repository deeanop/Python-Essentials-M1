```python
user = input("Username: ")
passw = input("Password: ")
special_char = "!@%"
while True:
    number_of_digits = 0
    number_of_specials = 0
    for i in range(0, len(passw)):
        if passw[i] in special_char:
            number_of_specials += 1
        if passw[i].isdigit():
            number_of_digits += 1
    if len(passw) < 7:
        print("The password is too short")
    elif number_of_specials == 0:
        print("The password should contain one of the following characters: %, !, @")
    elif number_of_specials > 1:
        print("The password should contain just one of the following characters: %, !, @")
    elif number_of_digits == 0:
        print("The password should include one digit")
    elif number_of_digits > 1:
        print("The password should include just one digit")
    elif passw[0] < 'A' or passw[0] > 'Z':
        print("The password should start with a capital")
    else:
        print("The password is accepted")
        break
    passw = input("Password: ")
    
```

```python
def putere(numar, exponent):
    return numar ** exponent
valoare = input("Introduceti un numar, sau q pentru iesire")
while True:
    if not valoare[0].isdigit():
        if valoare[0] == 'q':
            break
        else:
            print("Caracter invalid")
    elif valoare[0].isdigit():
        valoare_int = int(valoare)
        exponent = input("Introduceti exponentul")
        while True:
            if not exponent[0].isdigit():
                print("Caracter invalid")
            else:
                break
            exponent = input("Introduceti exponentul")
        exponent_int = int(exponent)
        print(putere(valoare_int, exponent_int))
    valoare = input("Introduceti un numar, sau q pentru iesire")
```

```python
def medie(varste: list):
    suma = 0
    for i in varste:
        suma += i
    return suma / len(varste)
lista_varste = list()
while True:
    try:
        numar = int(input("Introduceti un numar."))
        if numar <= 0:
            print("Numarul trebuie sa fie pozitiv")
            continue
        break
    except ValueError:
        print("Tip de date invalid")
for i in range(1, numar + 1):
    while True:
        try:
            varsta = int(input(f"Introduceti varsta utilizatorului {i}."))
            if varsta < 0:
                print("Varsta trebuie sa fie un numar pozitiv.")
                continue
            break
        except ValueError:
            print("Tip de date invalid")
    lista_varste.append(varsta)
print(medie(lista_varste))
```

```python
def suma(lista: list):
    sum = 0
    for i in lista:
        sum += i
    return sum
def medie(lista: list):
    sum = 0
    for i in lista:
        sum += i
    return sum / len(lista)
def putere(lista: list):
    for i in range(0, len(lista)):
        lista[i] = lista[i] ** 2
    return lista

meniu = {
    "1": medie,
    "2": suma,
    "3": putere,
}
lista_numere = list()
numar = input("Introduceti un numar.")
while True:
    if numar == 'z':
        break
    else:
        numar_float = float(numar)
        lista_numere.append(numar_float)
    numar = input("Introduceti un numar.")
print(meniu)
optiune = input("Introduceti optiunea")
if optiune == '1':
    print(suma(lista_numere))
elif optiune == '2':
    print(medie(lista_numere))
elif optiune == '3':
    print(putere(lista_numere))
```
